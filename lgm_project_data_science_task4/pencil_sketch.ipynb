```python
import cv2
import matplotlib.pyplot as plt
```

```python
#read the image
image=cv2.imread("kohli.jpg")
```

```python
#display the image
cv2.imshow(" Virat Kohli", image)

#display the image
cv2.waitKey(0)
plt.imshow(image)
```

```python
# create a new image by converting the original image to rgb format image
RGB_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2.imshow("New Kohli",RGB_image)
cv2.waitKey(0)
plt.imshow(RGB_image)
```

```python
# create a new image by converting the original image to greyscale image
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2.imshow("New Kohli",gray_image)
cv2.waitKey(0)
plt.imshow(gray_image)
```

```python
# invert the new greyscale image
inverted_image=255-gray_image
cv2.imshow("Inverted", inverted_image)
cv2.waitKey()
plt.imshow(inverted_image)
```

```python
# to blur the image by using the Gaussian Function in OpenCV
blurred=cv2.GaussianBlur(inverted_image,(21,21),0)
cv2.imshow("Blurred",blurred)
cv2.waitKey()
plt.imshow(blurred)
```

```python
#to invert the blurred image,then we can easily convert the image into a pencil sketch
inverted_blurred=255-blurred
pencil_sketch=cv2.divide(gray_image,inverted_image,scale=256.0)
cv2.imshow("Sketch",pencil_sketch)
cv2.waitKey(0)
plt.imshow(inverted_blurred)
```

```python
#looking at the original image and pencil sketch image
cv2.imshow("Original Kohli",image)
cv2.imshow("Pencil Sketch Kohli",pencil_sketch)
cv2.waitKey(0)
plt.imshow(image)
```
