```python
#virtually importing iris csv file
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris=load_iris()
dir(iris)
```

```python
# making a new column target which is our target variable
df['target']=iris.target
print(df)
```

```python
# showing/predicting the flower names corresponding to 0,1,2 target values
iris.target_names
```

```python
# printing dataframe for only those which haave target values as 0
df0=df[df.target==0]
```

```python
# printing dataframe for only those which haave target values as 1
df1=df[df.target==1]
```

```python
# printing dataframe for only those which haave target values as 2
df2=df[df.target==2]
```

```python
# printing dataframe for only those which have target values as 3 which in our case is not there
df3=df[df.target==3]
```

```python
# predicting/finding the flower_names based on the target_names values
# we have used lambda to define a function
df['flower_names']=df.target.apply(lambda x:iris.target_names[x])
print(df)
```

```python
# to draw a scatter diagram between sepal length and sepal width
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'])
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'])
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'])
# line of best fit
plt.plot(df['sepal length (cm)'],df['petal length (cm)'])
plt.show()
```

```python
# to draw a scatter diagram between petal length and petal width
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'])
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'])
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'])
plt.xlabel('petal length')
plt.ylabel('petal width')
# line of best fit
plt.plot(df['sepal width (cm)'],df['petal length (cm)'])
plt.show()
```

```python
# building the decision tree model
clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
```
